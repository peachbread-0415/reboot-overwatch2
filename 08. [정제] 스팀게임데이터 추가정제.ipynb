### 개발/배급사 누락값 크롤링

```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# 크롬 브라우저 설정
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_experimental_option('detach', True)
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

for id in df[df['배급사'].isna()]['appid']:
    url = f'https://store.steampowered.com/app/{id}/'
    driver.get(url)
    time.sleep(2)
    
    # 연령 확인 페이지가 있는 경우
    try:
        year_select = driver.find_element(By.ID, 'ageYear')
        Select(driver.find_element(By.ID, 'ageDay')).select_by_visible_text('1')
        Select(driver.find_element(By.ID, 'ageMonth')).select_by_visible_text('January')
        Select(driver.find_element(By.ID, 'ageYear')).select_by_visible_text('1995')
        time.sleep(0.5)

        view_button = driver.find_element(By.CSS_SELECTOR, 'a.btnv6_blue_hoverfade')
        view_button.click()
        time.sleep(1)
    except NoSuchElementException:
        # 연령 확인이 없는 경우는 그냥 넘어감
        pass

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 개발사 추출
    developer = 'No Developer'
    developer_div = soup.find('div', id='developers_list')
    if developer_div:
        a_tags = developer_div.find_all('a')
        if a_tags:
            developer = ', '.join([a.text.strip() for a in a_tags])
    
    # 배급사 추출
    publisher = 'No Publisher'
    publisher_label = soup.find('div', string='Publisher:')
    if publisher_label:
        parent = publisher_label.find_parent('div', class_='dev_row')
        if parent:
            summary_div = parent.find('div', class_='summary column')
            if summary_div:
                a_tags = summary_div.find_all('a')
                if a_tags:
                    publisher = ', '.join([a.text.strip() for a in a_tags])
    
    # DataFrame에 저장
    df.loc[df['appid'] == id, '배급사'] = publisher
    print(f'{id} 게임의 개발사: {developer}, 배급사: {publisher}')
```

```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# 크롬 브라우저 설정
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_experimental_option('detach', True)
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

for id in filtered_df['appid']:
    url = f'https://store.steampowered.com/app/{id}/'
    driver.get(url)
    time.sleep(2)

    # 연령 확인 페이지가 있는 경우
    try:
        year_select = driver.find_element(By.ID, 'ageYear')
        Select(driver.find_element(By.ID, 'ageDay')).select_by_visible_text('1')
        Select(driver.find_element(By.ID, 'ageMonth')).select_by_visible_text('January')
        Select(driver.find_element(By.ID, 'ageYear')).select_by_visible_text('1995')
        time.sleep(0.5)

        view_button = driver.find_element(By.CSS_SELECTOR, 'a.btnv6_blue_hoverfade')
        view_button.click()
        time.sleep(1)
    except NoSuchElementException:
        pass

    # 가격 정보 크롤링
    try:
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # 가격 정보 추출
        price_tag = soup.find('div', class_='game_purchase_price price')
        if price_tag:
            price = price_tag.text.strip()
        else:
            price = '0'
        
        df.loc[df['appid'] == id, '가격'] = price
        print(f'{id}의 가격: {price}')

    except Exception as e:
        print(f'{id} 처리 중 오류 발생: {e}')
        df.loc[df['appid'] == id, '가격'] = price
```
