# 게임 데이터와 접속자수 데이터 합침

```python
df = pd.read_csv('최종_스팀게임정보2차.csv')
```

```python
df2.info()
```

```python
df2
```

```python
df2.columns
```

```python
df.columns
```

```python
import pandas as pd

# 최근 3개월 컬럼 추출
# df2의 컬럼명 중 날짜 부분만 추출
date_cols = sorted(set(col.split('_')[0] for col in df2.columns if '_' in col and col != 'id'))

# 최근 3개월 날짜 코드 추출
last_3_months = date_cols[-3:]

# 붙일 컬럼 리스트
cols_to_merge = ['id'] + [f"{m}_Peak" for m in last_3_months] + [f"{m}_Avg" for m in last_3_months]

# 최근 3개월 데이터만 추출
df2_recent = df2[cols_to_merge].copy()

# df.appid와 df2.id를 기준으로 merge
df_merged = df.merge(df2_recent, left_on='appid', right_on='id', how='left')

# id 컬럼은 필요 없으면 제거
df_merged.drop(columns=['id'], inplace=True)

# 결과 확인
df_merged
```

```python
df_merged.info()
```

```python
cols = ['2505_Peak', '2506_Peak', '2507_Peak',
        '2505_Avg', '2506_Avg', '2507_Avg']

df_merged[cols] = df_merged[cols].fillna(0)
```

```python
df_merged.info()
```

```python
df_merged.to_csv('최종_스팀게임정보3차.csv', index=False)
```
