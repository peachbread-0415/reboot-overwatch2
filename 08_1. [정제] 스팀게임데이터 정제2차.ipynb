# 할인률 정제
- 게임명 중복 확인 -> 중복 제거
- 할인률, 할인기간 누락값 확인
- 할인률은 가격정보가 없거나 누락인 경우 전부 0으로 처리
- 열 순서 변경
- 할인기간 누락 or 데이터가 없는 칸은 x로 처리
- 할인기간을 시작과 종료 기간으로 나눔

```python
df['할인기간'] = df['할인기간'].apply(lambda x: 'x' if pd.isna(x) or x == '세일 정보가 없습니다.' else x)
df
```

```python
df['할인기간'].value_counts()
```

```python
df.info()
```

```python
# 시작기간과 종료기간을 나누는 함수
def split_discount_period(period):
    period = period.strip()
    if period == 'x':
        return pd.Series(['x', 'x'])
    elif '~' in period:
        parts = period.split('~')
        start = parts[0].strip()
        end = parts[1].strip() if len(parts) > 1 and parts[1].strip() else '~'
        return pd.Series([start, end])
    else:
        return pd.Series(['x', 'x'])

# 새로운 열로 분리
df[['할인시작', '할인종료']] = df['할인기간'].apply(split_discount_period)
df
```

```python
df['할인종료'].value_counts()
```

```python
df.to_csv('최종_할인률1차.csv', index=False)
```

```python
# 숫자만 추출하는 함수
def extract_number(text):
    import re
    if isinstance(text, str):
        match = re.search(r'\d+\.?\d*', text)  # 숫자 또는 소수점 숫자 추출
        if match:
            return float(match.group())
    return None  # 숫자 없으면 None 반환

# 적용
asd['play_time_num'] = asd['play_time'].apply(extract_number)
asd['recently_time_num'] = asd['recently_time'].apply(extract_number)

print(asd)
```

```python
# 먼저 df1을 딕셔너리로 변환 (appid를 키로, 나머지 열들을 서브 딕셔너리로)
# appid 기준으로 중복 제거 (첫 번째 값만 유지)
df1_unique = df1.drop_duplicates(subset='appid', keep='first')

# 이제 딕셔너리로 변환 가능
df1_dict = df1_unique.set_index('appid').to_dict(orient='index')

# 각 행마다 appid를 기준으로 df1_dict에서 정보 가져오기
def map_row(row):
    extra_info = df1_dict.get(row['appid'], {})
    for key, value in extra_info.items():
        row[key] = value
    return row

# 적용
df = df.apply(map_row, axis=1)
```
